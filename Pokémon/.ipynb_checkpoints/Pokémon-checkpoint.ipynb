| **habitat** | 서식지 | 자연에서 사는 곳 |

| **hatch_counter** | 부화 카운터 | 알이 부화하는 데 걸리는 걸음 수 |
| **gender_rate** | 성별 비율 | 암수 비율 |
| **description** | 설명 | 도감 설명 |
| **sprite_url** | 이미지 URL | 포켓몬 도트 이미지 링크 |
| **is_dual_type** | 이중타입 여부 | 타입이 2개인지 |
| **bmi** | BMI | 몸무게/키² |
| **attack_defense_ratio** | 공격/방어 비율 | 공격력 대비 방어력 |

| **special_total** | 특수 합계 | sp_attack + sp_defense |

| **defensive_total** | 방어 합계 | defense + sp_defense |
| **gender_distribution** | 성별 분포 | 성별 비율 텍스트 |
| **stat_tier** | 스탯 등급 | 종족값 기준 등급 |

### **체력 스탯, 물리 공격 스탯, 물리 방어 스탯, 특수 공격 스탯, 특수 방어 스탯, 속도 스탯으로 전설 포켓몬인지 아닌지 예측하기**

#### 트러블 슈팅
1. 초기에 포켓몬 이름을 예측 목표로 설정했으나,
name은 범주형 변수이고 동일한 타입과 등급을 가진 포켓몬이 
다수 존재해 예측이 불가능함을 확인

→ **수치형 타깃(is_legendary)으로 전환하여 분류 문제로 재정의**

2. 이진 분류형식의 정답 label을 선형 회귀로 예측 시도하였으나
model.score가 0.1 이라는 아주 낮은 수치가 기록됨
np.log1p를 사용해 min과 max의 편차도 줄여보았지만 큰 차이가 없었음 

→ **모델을 이진 분류모델(KNN)로 교체하여 재학습 시켜보니 정확도 0.9**

```python
X_TrainAndTestData = df[['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']]
```

```python
X_TrainAndTestData
```

```python
df['is_legendary']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np

X = X_TrainAndTestData
y = pd.DataFrame(df['is_legendary'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

X.describe()
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
print(f'{model.score(X_test, y_test):.4f}')
```
