### **체력 스탯, 물리 공격 스탯, 물리 방어 스탯, 특수 공격 스탯, 특수 방어 스탯, 속도 스탯으로 전설 포켓몬인지 아닌지 예측하기**

```python
X_TrainAndTestData = df[['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']]
```

```python
X_TrainAndTestData
```

```python
df['is_legendary']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np

X = X_TrainAndTestData
y = pd.DataFrame(df['is_legendary'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)

scaler = StandardScaler()
scaler.fit(X_train)
X_train_s = scaler.transform(X_train)
X_test_s = scaler.transform(X_test)

X.describe()
```

```python
from sklearn.svm import SVC
model = SVC(class_weight = 'balanced')
model.fit(X_train_s, y_train)
```

```python
y_pred = model.predict(X_test_s)
y_pred
```

```python
print(f'{model.score(X_test, y_test):.4f}')
```

```python
pd.Series(y_pred).value_counts()
```

```python
from sklearn.metrics import f1_score
print(f1_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize = (6, 4))
sns.heatmap(cm, annot = True, fmt = 'd', xticklabels = ['normal', 'legendary'], yticklabels = ['normal', 'legendary'])

plt.title('SVM Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

### 2. 스케일링 누락

문제: KNN은 유클리드 거리 기반인데 스케일링 없이 학습 → 값이 큰 피처가 거리 계산 지배

해결: StandardScaler 적용, train 기준으로 fit 후 test는 transform만

### 3. 클래스 불균형

해결: KNN 대신 SVC(class_weight='balanced') 적용 → F1 0.5에서 0.6으로 개선
