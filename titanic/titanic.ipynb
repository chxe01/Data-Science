```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df['Embarked'].mode()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Age'] = df['Age'].astype(int)
df
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df
```

```python
df = df.drop('Cabin', axis = 1)
df
```

```python
df.isnull().sum()
```

```python
male = df[df['Sex'] == 'male']
female = df[df['Sex'] == 'female']

male = male['Survived'].mean() * 100
female = female['Survived'].mean() * 100

male, female
```

```python
color = ['#283618', '#606c38']
plt.title('Survival rates by Sex')
plt.bar(['Male', 'Female'], [male, female], color = color)
plt.show()
```

#### 분석 결과
남성 승객 생존율은 약 18.8%, 여성 승객 생존율은 약 74.2%로
여성 승객이 남성 승객보다 생존율이 약 4배 가량 높다.

타이타닉 침몰 당시 'Women and children first(여성과 어린이 먼저)' 원칙으로
승객을 구조했기 때문에 여성 생존율이 압도적으로 높게 나타난 것으로 볼 수 있다.

```python
S = df[df['Embarked'] == 'S']
Q = df[df['Embarked'] == 'Q']
C = df[df['Embarked'] == 'C']

S = S['Survived'].mean() * 100
Q = Q['Survived'].mean() * 100
C = C['Survived'].mean() * 100

print (f'S: {S:.4f}')
print (f'Q: {Q:.4f}')
print (f'C: {C:.4f}')
```

```python
color = ['#90e0ef', '#0096c7', '#023e8a']
plt.title('Survival rates by Embarked')
plt.bar(['S', 'Q', 'C'], [S, Q, C], color = color)
plt.show()
```

```python
_1 = df[df['Pclass'] == 1]
_2 = df[df['Pclass'] == 2]
_3 = df[df['Pclass'] == 3]

_1 = _1['Survived'].mean() * 100
_2 = _2['Survived'].mean() * 100
_3 = _3['Survived'].mean() * 100

print (f'1: {_1:.4f}')
print (f'2: {_2:.4f}')
print (f'3: {_3:.4f}')
```

```python
color = ['#90e0ef', '#0096c7', '#023e8a']
plt.title('Survival rates by Pclass')
plt.bar(['1', '2', '3'], [_1, _2, _3], color = color)
plt.show()
```

### 분석 결과
각 등급별 생존율은 1등급 62.9  2등급 47.2  3등급 24.2 으로, 1등급 -> 2등급 -> 3등급 순으로 생존율이 낮은 걸 볼 수 있다.

타이타닉 침몰 당시 1등급석 승객을 먼저 구출하기도 하였고, 3등급석은 배 아래쪽, 1등급석은 배 위쪽에 있어 물이 먼저 찼을 3등급석 승객의 생존율은 낮은 걸 알 수 있다.
